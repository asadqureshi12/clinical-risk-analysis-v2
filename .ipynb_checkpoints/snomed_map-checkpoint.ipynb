### SNOMED Conditions Processing Script (Defining the run function)

This script defines the `run()` function and supporting utilities to process Synthea condition data into patient-level tables.

- **`load_data()`** – Reads the raw `conditions.csv` and the SNOMED universe mapping.
- **`map_conditions()`** – Merges the raw `conditions.csv` with the SNOMED universe mapping to assign a **CATEGORY** to each condition.
- **`build_wide()`** – Produces a **wide-format table**: one row per patient with binary flags for each category and a `category_count`.
- **`build_long()`** – Produces a **long-format table**: one row per patient-category pair.
- **`run()`** – Orchestrates the above steps, saves CSV outputs in the `outputs/` folder, and prints summary statistics.

**Outputs saved:**
- `patient_conditions_wide.csv`
- `patient_conditions_long.csv`

```python
%%writefile src/build_snomed_universe.py
import pandas as pd
import os

def load_data(conditions_path, universe_path):
    conditions = pd.read_csv(conditions_path)
    universe   = pd.read_csv(universe_path)
    return conditions, universe

def map_conditions(conditions, universe):
    merged = conditions.merge(
        universe[["CODE", "CATEGORY"]],
        on="CODE",
        how="left"
    )
    total     = len(conditions)
    matched   = merged["CATEGORY"].notna().sum()
    unmatched = total - matched
    print(f"[snomed_conditions] Condition rows total     : {total}")
    print(f"[snomed_conditions] Matched to category      : {matched} ({matched/total*100:.1f}%)")
    print(f"[snomed_conditions] Unmatched (out of scope) : {unmatched} ({unmatched/total*100:.1f}%)")
    merged["CATEGORY"] = merged["CATEGORY"].fillna("OTHER")
    return merged

def build_wide(merged):
    wide = (
        merged.groupby(["PATIENT", "CATEGORY"])
        .size()
        .unstack(fill_value=0)
        .clip(upper=1)
        .reset_index()
    )
    wide["category_count"] = wide.drop(columns="PATIENT").sum(axis=1)
    return wide

def build_long(merged):
    long = (
        merged[["PATIENT", "CATEGORY"]]
        .drop_duplicates()
        .sort_values(["PATIENT", "CATEGORY"])
        .reset_index(drop=True)
    )
    return long

def run(conditions_path, universe_path, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)
    conditions, universe = load_data(conditions_path, universe_path)
    merged               = map_conditions(conditions, universe)
    wide                 = build_wide(merged)
    long                 = build_long(merged)
    wide_path = os.path.join(output_dir, "patient_conditions_wide.csv")
    long_path = os.path.join(output_dir, "patient_conditions_long.csv")
    wide.to_csv(wide_path, index=False)
    long.to_csv(long_path, index=False)
    print(f"\n[snomed_conditions] Wide format saved : {wide_path}")
    print(f"[snomed_conditions] Long format saved : {long_path}")
    print(f"[snomed_conditions] Patients in wide  : {len(wide)}")
    print(f"[snomed_conditions] Rows in long      : {len(long)}")
    return wide, long, merged
```

# Sprint 2 — Clinical Risk Scoring
**Date:** 2026-03-31  
**Purpose:** Assign evidence-based risk scores to each patient using a multi-layered scoring engine. Layers include category weights aligned with NHS Long Term Plan priorities, age-stratified adjustments, multimorbidity tiers, polypharmacy, utilisation, and recurrence bonuses. Each patient is classified into one of four risk tiers: Low, Medium, High, or Critical.

## Scoring Methodology Note

Patient risk scores were assigned using a **percentile-based system** derived from the current dataset. This approach ensures that segmentation reflects the **actual distribution of patient risk** in this cohort, highlighting high-risk groups effectively.

Thresholds may differ for other populations depending on their risk distributions.

```python
import pandas as pd

conditions  = pd.read_csv("data/conditions.csv")
patients    = pd.read_csv("data/patients.csv")
medications = pd.read_csv("data/medications.csv")
encounters  = pd.read_csv("data/encounters.csv")

print(f"conditions  : {len(conditions)} rows")
print(f"patients    : {len(patients)} rows")
print(f"medications : {len(medications)} rows")
print(f"encounters  : {len(encounters)} rows")
```

## Calculating med cap
- Medication counts capped at 100 — values above this threshold are 
clinically implausible as concurrent prescriptions and are attributable 
to cumulative historical records in the Synthea dataset. The median 
medication count in the cohort is 8, with a 99th percentile of 420 
and a maximum of 3,313 prior to capping.

```python
med_count = medications.groupby("PATIENT").size()
med_count = med_count.sort_values()

percentile_99 = med_count.quantile(0.99)
percentile_95 = med_count.quantile(0.95)
percentile_90 = med_count.quantile(0.90)

print(f"90th percentile : {percentile_90:.0f}")
print(f"95th percentile : {percentile_95:.0f}")
print(f"99th percentile : {percentile_99:.0f}")
print(f"Maximum value   : {med_count.max()}")
print(f"Mean            : {med_count.mean():.1f}")
print(f"Median          : {med_count.median():.0f}")
```

## Step 1 — Define Risk Weights and Parameters
All scoring parameters are defined in one place. Weights are aligned with NHS Long Term Plan priorities and NICE clinical burden guidance. Adjust values here only — no changes needed in downstream cells.

```python
CATEGORY_WEIGHTS = {
    "DIABETES & COMPLICATIONS":             4,
    "CARDIOVASCULAR":                       4,
    "ENDOCRINOLOGY":                        3,
    "NEUROLOGY / CEREBROVASCULAR":          3,
    "NEPHROLOGY & ELECTROLYTES DISORDERS":  3,
    "PULMONOLOGY":                          2,
    "PSYCHIATRY":                           2,
    "DRUG INTERACTIONS / ADDICTION":        2,
    "GENERAL SURGERY":                      1,
    "ORTHOPEDICS & RHEUMATOLOGY":           1,
    "GYNECOLOGY & OBSTETRICS":              1,
    "TRAUMA":                               0.5,
    "MALE REPRODUCTIVE":                    0.5,
    "ENT":                                  0.5,
    "OTHER":                                0.5,
}

MULTIMORBIDITY_TIERS = {
    2: 1,
    5: 2,
    8: 3,
}

AGE_SENSITIVE_CATEGORIES = {
    "DIABETES & COMPLICATIONS",
    "CARDIOVASCULAR",
    "NEUROLOGY / CEREBROVASCULAR",
    "ENDOCRINOLOGY",
    "NEPHROLOGY & ELECTROLYTES DISORDERS",
}

AGE_MULTIPLIERS = {
    "0-40":  1.0,
    "41-65": 1.25,
    "66+":   1.5,
}

POLYPHARMACY_THRESHOLDS = {
    25:  1,
    50:  2,
    100: 3,
}

HIGH_UTILISATION_THRESHOLDS = {
    20:  1,
    50:  2,
    100: 3,
}

RECURRENCE_BONUS_THRESHOLD = 3
RECURRENCE_BONUS_SCORE     = 0.5
MEDICATION_OUTLIER_CAP     = 100

print("Parameters loaded.")
```

## Step 2: Recurrence Feature Engineering (Encounter Frequency per Category)

This cell calculates the number of encounters each patient has within each clinical category and integrates this information into the main scoring dataset.

First, the data is grouped by `PATIENT` and `CATEGORY`, and the total number of occurrences is counted. This represents how many times a patient has interacted with the healthcare system for a given condition.

The result is then reshaped into a wide format where each category becomes a separate column. Missing values are filled with 0, indicating no encounters in that category.

All category columns are prefixed with `recurrence_` to distinguish them from the binary presence indicators (0/1) used in the base score.

Finally, the recurrence table is merged with the main `scored` dataset, ensuring each patient now has both:
- Binary indicators for condition presence
- Numerical counts representing encounter frequency per category

These recurrence features are later used to apply recurrence-based bonuses, capturing repeated healthcare utilisation in clinically significant categories.

```python
recurrence = (
    merged.groupby(["PATIENT", "CATEGORY"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

recurrence.columns = (
    ["PATIENT"] +
    [f"recurrence_{col}" for col in recurrence.columns if col != "PATIENT"]
)

scored = wide.merge(recurrence, on="PATIENT", how="left")

print(f"Recurrence columns added : {[c for c in scored.columns if c.startswith('recurrence_')][:3]} ...")
print(f"Patients in scored table : {len(scored)}")
```

## Step 3: Base Score
Vectorised dot product of binary category flags and NHS-aligned risk weights.

```python
category_cols = [col for col in scored.columns if col in CATEGORY_WEIGHTS]

weight_vector = pd.Series(
    {col: CATEGORY_WEIGHTS[col] for col in category_cols}
)

scored["base_score"] = scored[category_cols].dot(weight_vector)

print(f"Base score range : {scored['base_score'].min()} – {scored['base_score'].max()}")
scored[["PATIENT", "base_score"]].head()
```

## Step 4: Multimorbidity Tier Bonus
 Patients with conditions spanning multiple clinical categories receive a tiered bonus reflecting the compounding burden of multimorbidity. Thresholds are set at ≥2, ≥5, and ≥8 distinct categories, awarding a bonus of 1, 2, and 3 respectively. This layer is category-agnostic — it captures breadth of disease burden independently of which specific conditions are present.

```python
def multimorbidity_tier_bonus(category_count):
    bonus = 0
    for threshold, score in sorted(MULTIMORBIDITY_TIERS.items()):
        if category_count >= threshold:
            bonus = score
    return bonus

scored["multimorbidity_tier_bonus"] = scored["category_count"].apply(multimorbidity_tier_bonus)

print(f"Multimorbidity tier distribution :")
print(scored["multimorbidity_tier_bonus"].value_counts().sort_index())
print(f"\nPatients with ≥2 categories  : {(scored['category_count'] >= 2).sum()}")
print(f"Patients with ≥5 categories  : {(scored['category_count'] >= 5).sum()}")
print(f"Patients with ≥8 categories  : {(scored['category_count'] >= 8).sum()}")
```

## Step 5: Deciding Reference Date for Age Calculation
Age is calculated relative to 2020-04-25 — the latest condition record date 
in the Synthea dataset. This represents the end of the observation window 
and ensures age values are consistent with the clinical data period.

```python
import pandas as pd

conditions = pd.read_csv("data/conditions.csv")
conditions["START"] = pd.to_datetime(conditions["START"])
print("Earliest date:", conditions["START"].min())
print("Latest date :", conditions["START"].max())
```

## Step 6: Age Calculation and Age-Stratified Adjustment
High-burden categories are upweighted for older patients, reflecting increased 
clinical risk with age. Age bands: 0–40 (no adjustment), 41–65 (×1.25), 66+ (×1.5).

```python
AGE_REFERENCE_DATE = pd.Timestamp("2020-04-25")

patients_clean = patients.copy()
patients_clean["age"] = (
    AGE_REFERENCE_DATE - pd.to_datetime(patients_clean["BIRTHDATE"])
).dt.days // 365

def get_age_band(age):
    if age <= 40:
        return "0-40"
    elif age <= 65:
        return "41-65"
    else:
        return "66+"

patients_clean["age_band"] = patients_clean["age"].apply(get_age_band)

scored = scored.merge(
    patients_clean[["Id", "age", "age_band"]].rename(columns={"Id": "PATIENT"}),
    on="PATIENT",
    how="left"
)

age_adjustment = pd.Series(0.0, index=scored.index)

for col in AGE_SENSITIVE_CATEGORIES:
    if col in scored.columns and col in CATEGORY_WEIGHTS:
        base_weight = CATEGORY_WEIGHTS[col]
        for band, multiplier in AGE_MULTIPLIERS.items():
            mask = (scored["age_band"] == band) & (scored[col] == 1)
            age_adjustment[mask] += base_weight * (multiplier - 1.0)

scored["age_adjustment"] = age_adjustment.round(2)

print(f"Age band distribution :")
print(scored["age_band"].value_counts())
```

## Step 7: Utilisation and Polypharmacy
Medication counts are capped at 100 prior to scoring. Analysis shows 87 patients 
(7.9% of the cohort) exceed this threshold, with a maximum raw value of 3,313 — 
attributable to cumulative historical prescriptions in the Synthea dataset rather 
than concurrent medication burden. The median medication count is 8. Capping at 100 
removes clinically implausible values whilst preserving the distribution for 
the remaining 92.1% of patients. In a production NHS system, only active medications 
at a defined reference date would be used.

```python
med_counts = (
    medications.groupby("PATIENT")
    .size()
    .reset_index(name="medication_count")
)
med_counts["medication_count"] = med_counts["medication_count"].clip(upper=MEDICATION_OUTLIER_CAP)

enc_counts = (
    encounters.groupby("PATIENT")
    .size()
    .reset_index(name="encounter_count")
)

scored = scored.merge(med_counts, on="PATIENT", how="left")
scored = scored.merge(enc_counts, on="PATIENT", how="left")
scored["medication_count"] = scored["medication_count"].fillna(0)
scored["encounter_count"]  = scored["encounter_count"].fillna(0)

def polypharmacy_bonus(med_count):
    bonus = 0
    for threshold, score in sorted(POLYPHARMACY_THRESHOLDS.items()):
        if med_count >= threshold:
            bonus = score
    return bonus

def utilisation_bonus(enc_count):
    bonus = 0
    for threshold, score in sorted(HIGH_UTILISATION_THRESHOLDS.items()):
        if enc_count >= threshold:
            bonus = score
    return bonus

scored["polypharmacy_bonus"] = scored["medication_count"].apply(polypharmacy_bonus)
scored["utilisation_bonus"]  = scored["encounter_count"].apply(utilisation_bonus)

print(f"Patients with polypharmacy bonus   : {(scored['polypharmacy_bonus'] > 0).sum()}")
print(f"Patients with utilisation bonus    : {(scored['utilisation_bonus'] > 0).sum()}")
print(f"Medication count (capped) — max    : {scored['medication_count'].max()}")
print(f"Encounter count — max              : {scored['encounter_count'].max()}")
print(f"Polypharmacy bonus distribution    :")
print(scored["polypharmacy_bonus"].value_counts().sort_index())
print(f"Utilisation bonus distribution     :")
print(scored["utilisation_bonus"].value_counts().sort_index())
```

## Step 8: Applying recurrence bonus to age sensitive groups

```python
for col in AGE_SENSITIVE_CATEGORIES:
    rec_col = f"recurrence_{col}"
    if rec_col in scored.columns:
        scored[f"{col}_rec_bonus"] = (scored[rec_col] >= RECURRENCE_BONUS_THRESHOLD) * RECURRENCE_BONUS_SCORE
```

## Step 9 — Risk Score Calculation
Aggregates all scoring layers into a single risk score per patient. Recurrence bonus columns are summed across qualifying age-sensitive categories before the final score is computed.

```python
recurrence_bonus_cols = [col for col in scored.columns if col.endswith("_rec_bonus")]

scored["recurrence_bonus"] = scored[recurrence_bonus_cols].sum(axis=1)

scored["risk_score"] = (
    scored["base_score"] +
    scored["multimorbidity_tier_bonus"] +
    scored["age_adjustment"] +
    scored["polypharmacy_bonus"] +
    scored["utilisation_bonus"] +
    scored["recurrence_bonus"]
).round(2)

print(f"Risk scores calculated for {len(scored)} patients.")
```

## Step 10 — Score Distribution Analysis and Risk Tier Calculation

```python
print("── Score Distribution ───────────────────────────────")
print(scored["risk_score"].describe())
print()
print(f"25th percentile  : {scored['risk_score'].quantile(0.25)}")
print(f"50th percentile  : {scored['risk_score'].quantile(0.50)}")
print(f"75th percentile  : {scored['risk_score'].quantile(0.75)}")
print(f"90th percentile  : {scored['risk_score'].quantile(0.90)}")
print(f"95th percentile  : {scored['risk_score'].quantile(0.95)}")
```

## Step 11 — Risk Tier Assignment
Assigns each patient to one of four risk tiers based on the observed score distribution. Thresholds are set at rounded percentile values: Critical (≥27, ~95th), High (≥17, ~75th), Medium (≥5, ~25th), Low (below 25th).

```python
def assign_tier(score):
    if score >= 27:
        return "Critical"
    elif score >= 17:
        return "High"
    elif score >= 5:
        return "Medium"
    else:
        return "Low"

scored["risk_tier"] = scored["risk_score"].apply(assign_tier)
print(scored["risk_tier"].value_counts())
```

## Risk Scorer Module Export
Consolidates all scoring logic into `src/risk_scorer.py` for pipeline use.

```python
%%writefile src/compute_risk_scores.py
import pandas as pd
import os

CATEGORY_WEIGHTS = {
    "DIABETES & COMPLICATIONS":             4,
    "CARDIOVASCULAR":                       4,
    "ENDOCRINOLOGY":                        3,
    "NEUROLOGY / CEREBROVASCULAR":          3,
    "NEPHROLOGY & ELECTROLYTES DISORDERS":  3,
    "PULMONOLOGY":                          2,
    "PSYCHIATRY":                           2,
    "DRUG INTERACTIONS / ADDICTION":        2,
    "GENERAL SURGERY":                      1,
    "ORTHOPEDICS & RHEUMATOLOGY":           1,
    "GYNECOLOGY & OBSTETRICS":              1,
    "TRAUMA":                               0.5,
    "MALE REPRODUCTIVE":                    0.5,
    "ENT":                                  0.5,
    "OTHER":                                0.5,
}

MULTIMORBIDITY_TIERS = {
    2: 1,
    5: 2,
    8: 3,
}

AGE_SENSITIVE_CATEGORIES = {
    "DIABETES & COMPLICATIONS",
    "CARDIOVASCULAR",
    "NEUROLOGY / CEREBROVASCULAR",
    "ENDOCRINOLOGY",
    "NEPHROLOGY & ELECTROLYTES DISORDERS",
}

AGE_MULTIPLIERS = {
    "0-40":  1.0,
    "41-65": 1.25,
    "66+":   1.5,
}

POLYPHARMACY_THRESHOLDS = {
    25:  1,
    50:  2,
    100: 3,
}

HIGH_UTILISATION_THRESHOLDS = {
    20:  1,
    50:  2,
    100: 3,
}

RECURRENCE_BONUS_THRESHOLD = 3
RECURRENCE_BONUS_SCORE     = 0.5
MEDICATION_OUTLIER_CAP     = 100
AGE_REFERENCE_DATE         = pd.Timestamp("2020-04-25")


def load_data(conditions_path, patients_path, medications_path, encounters_path):
    conditions  = pd.read_csv(conditions_path)
    patients    = pd.read_csv(patients_path)
    medications = pd.read_csv(medications_path)
    encounters  = pd.read_csv(encounters_path)
    return conditions, patients, medications, encounters


def build_scored_base(wide, merged):
    recurrence = (
        merged.groupby(["PATIENT", "CATEGORY"])
        .size()
        .unstack(fill_value=0)
        .reset_index()
    )
    recurrence.columns = (
        ["PATIENT"] +
        [f"recurrence_{col}" for col in recurrence.columns if col != "PATIENT"]
    )
    scored = wide.merge(recurrence, on="PATIENT", how="left")
    return scored


def score_base(scored):
    category_cols = [col for col in scored.columns if col in CATEGORY_WEIGHTS]
    weight_vector = pd.Series({col: CATEGORY_WEIGHTS[col] for col in category_cols})
    scored["base_score"] = scored[category_cols].dot(weight_vector)
    return scored


def score_multimorbidity(scored):
    def multimorbidity_tier_bonus(category_count):
        bonus = 0
        for threshold, score in sorted(MULTIMORBIDITY_TIERS.items()):
            if category_count >= threshold:
                bonus = score
        return bonus
    scored["multimorbidity_tier_bonus"] = scored["category_count"].apply(multimorbidity_tier_bonus)
    return scored


def score_age(scored, patients):
    patients_clean = patients.copy()
    patients_clean["age"] = (
        AGE_REFERENCE_DATE - pd.to_datetime(patients_clean["BIRTHDATE"])
    ).dt.days // 365

    def get_age_band(age):
        if age <= 40:
            return "0-40"
        elif age <= 65:
            return "41-65"
        else:
            return "66+"

    patients_clean["age_band"] = patients_clean["age"].apply(get_age_band)
    scored = scored.merge(
        patients_clean[["Id", "age", "age_band"]].rename(columns={"Id": "PATIENT"}),
        on="PATIENT",
        how="left"
    )
    age_adjustment = pd.Series(0.0, index=scored.index)
    for col in AGE_SENSITIVE_CATEGORIES:
        if col in scored.columns and col in CATEGORY_WEIGHTS:
            base_weight = CATEGORY_WEIGHTS[col]
            for band, multiplier in AGE_MULTIPLIERS.items():
                mask = (scored["age_band"] == band) & (scored[col] == 1)
                age_adjustment[mask] += base_weight * (multiplier - 1.0)
    scored["age_adjustment"] = age_adjustment.round(2)
    return scored


def score_polypharmacy_utilisation(scored, medications, encounters):
    med_counts = (
        medications.groupby("PATIENT")
        .size()
        .reset_index(name="medication_count")
    )
    med_counts["medication_count"] = med_counts["medication_count"].clip(upper=MEDICATION_OUTLIER_CAP)
    enc_counts = (
        encounters.groupby("PATIENT")
        .size()
        .reset_index(name="encounter_count")
    )
    scored = scored.merge(med_counts, on="PATIENT", how="left")
    scored = scored.merge(enc_counts, on="PATIENT", how="left")
    scored["medication_count"] = scored["medication_count"].fillna(0)
    scored["encounter_count"]  = scored["encounter_count"].fillna(0)

    def polypharmacy_bonus(med_count):
        bonus = 0
        for threshold, score in sorted(POLYPHARMACY_THRESHOLDS.items()):
            if med_count >= threshold:
                bonus = score
        return bonus

    def utilisation_bonus(enc_count):
        bonus = 0
        for threshold, score in sorted(HIGH_UTILISATION_THRESHOLDS.items()):
            if enc_count >= threshold:
                bonus = score
        return bonus

    scored["polypharmacy_bonus"] = scored["medication_count"].apply(polypharmacy_bonus)
    scored["utilisation_bonus"]  = scored["encounter_count"].apply(utilisation_bonus)
    return scored


def score_recurrence(scored):
    for col in AGE_SENSITIVE_CATEGORIES:
        rec_col = f"recurrence_{col}"
        if rec_col in scored.columns:
            scored[f"{col}_rec_bonus"] = (scored[rec_col] >= RECURRENCE_BONUS_THRESHOLD) * RECURRENCE_BONUS_SCORE
    recurrence_bonus_cols = [col for col in scored.columns if col.endswith("_rec_bonus")]
    scored["recurrence_bonus"] = scored[recurrence_bonus_cols].sum(axis=1)
    return scored


def assign_tier(score):
    if score >= 27:
        return "Critical"
    elif score >= 17:
        return "High"
    elif score >= 5:
        return "Medium"
    else:
        return "Low"


def calculate_final_score(scored):
    scored["risk_score"] = (
        scored["base_score"] +
        scored["multimorbidity_tier_bonus"] +
        scored["age_adjustment"] +
        scored["polypharmacy_bonus"] +
        scored["utilisation_bonus"] +
        scored["recurrence_bonus"]
    ).round(2)
    scored["risk_tier"] = scored["risk_score"].apply(assign_tier)
    return scored


def run(wide, merged, patients, medications, encounters, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)

    scored = build_scored_base(wide, merged)
    scored = score_base(scored)
    scored = score_multimorbidity(scored)
    scored = score_age(scored, patients)
    scored = score_polypharmacy_utilisation(scored, medications, encounters)
    scored = score_recurrence(scored)
    scored = calculate_final_score(scored)

    print(f"[risk_scorer] Patients scored        : {len(scored)}")
    print(f"[risk_scorer] Risk score range       : {scored['risk_score'].min()} – {scored['risk_score'].max()}")
    print(f"[risk_scorer] Mean score             : {scored['risk_score'].mean():.2f}")
    print()

    tier_counts = scored["risk_tier"].value_counts().reindex(["Critical", "High", "Medium", "Low"])
    tier_pct    = (tier_counts / len(scored) * 100).round(1)
    for tier in ["Critical", "High", "Medium", "Low"]:
        print(f"[risk_scorer] {tier:<10} : {tier_counts[tier]:>4} patients ({tier_pct[tier]}%)")

    output_path = os.path.join(output_dir, "patient_risk_scores.csv")
    scored.to_csv(output_path, index=False)
    print(f"\n[risk_scorer] Exported : {output_path}")

    return scored
```

# Sprint 3 — Patient Segmentation
**Date:** 2026-04-01  
**Purpose:** Assign each patient to clinically meaningful segments based on condition profile, demographics, risk tier, and utilisation. Each patient receives a primary segment (single best-fit) and a full list of all qualifying segments for research and analytics use.

## Step 1 — Load Acute SNOMED Codes
Acute condition codes are loaded from the versioned SNOMED universe reference table 
(snomed_conditions_universe.csv) using the is_acute flag. This separates clinical 
classification decisions from analysis code — any update to the acute code list is 
made in the reference table without modifying the pipeline.

```python
universe = pd.read_csv("outputs/snomed_conditions_universe.csv")
ACUTE_SNOMED_CODES = set(universe[universe["is_acute"] == 1]["CODE"])
print(f"Acute SNOMED codes loaded from universe : {len(ACUTE_SNOMED_CODES)}")
```

## Step 2 — Derive Acute Condition Count and Overlay Flag
Cross-references each patient's condition history against the acute SNOMED code list and counts the number of distinct acute condition types per patient. Patients with five or more distinct acute conditions receive an ACUTE_OVERLAY flag of 1, indicating a recurring pattern of acute presentations across multiple clinical domains.

```python
acute_condition_counts = (
    conditions[conditions["CODE"].isin(ACUTE_SNOMED_CODES)]
    .groupby("PATIENT")["CODE"]
    .nunique()
    .reset_index(name="acute_condition_count")
)

scored = scored.merge(acute_condition_counts, on="PATIENT", how="left")
scored["acute_condition_count"] = scored["acute_condition_count"].fillna(0)

scored["ACUTE_OVERLAY"] = (scored["acute_condition_count"] >= 5).astype(int)

print(f"Acute Overlay eligible patients : {scored['ACUTE_OVERLAY'].sum()}")
print(f"As % of cohort                  : {scored['ACUTE_OVERLAY'].mean()*100:.1f}%")
```

## Step 3 — Derive Cardio-Metabolic Flag
Identifies patients with cardiovascular disease alongside diabetes complications. This combination represents the classic NHS priority cohort for long-term condition management and cardiovascular risk reduction.

```python
scored["CARDIO_METABOLIC"] = (
    (scored["CARDIOVASCULAR"] == 1) &
    (scored["DIABETES & COMPLICATIONS"] == 1)
).astype(int)

print(f"Cardio-Metabolic eligible patients : {scored['CARDIO_METABOLIC'].sum()}")
print(f"As % of cohort                     : {scored['CARDIO_METABOLIC'].mean()*100:.1f}%")
```

## Step 4 — Multimorbid Frail Segment

This segment refines the previous Frail Elderly definition by incorporating multimorbidity:

- Patients aged ≥66 with ≥3 condition categories
### Rationale

- Age alone is already included in the base risk score  
- Adding multimorbidity identifies patients with compounded clinical complexity  
- Better reflects real-world NHS cohorts where frailty is linked to multiple chronic conditions, not age alone

```python
scored["MULTIMORBID_FRAIL"] = (
    (scored["age"] >= 66) &
    (scored["category_count"] >= 3) &
    (scored["CARDIO_METABOLIC"] == 0)
).astype(int)

print(f"Multimorbid Frail eligible patients : {scored['MULTIMORBID_FRAIL'].sum()}")
print(f"As % of cohort                     : {scored['MULTIMORBID_FRAIL'].mean()*100:.1f}%")
```

## Step 5 — Midlife Escalation Refinement

Midlife Escalation is defined as patients aged 41–65 with ≥50 healthcare encounters, combining age-based risk with utilisation behaviour.

### Rationale

- Moves beyond age alone to identify patients with active healthcare demand  
- Captures a preventative cohort where intervention can reduce progression to frailty  
- Reflects both clinical risk and system interaction, improving actionability

```python
scored["MIDLIFE_ESCALATION"] = (
    (scored["age_band"] == "41-65") &
    (scored["encounter_count"] > 50)
).astype(int)

print(f"Midlife Escalation eligible patients : {scored['MIDLIFE_ESCALATION'].sum()}")
print(f"As % of cohort                       : {scored['MIDLIFE_ESCALATION'].mean()*100:.1f}%")
```

## Step 7 — Multi-Segment Overlap Analysis
Counts how many segments each patient qualifies for before primary segment assignment. This informs the priority hierarchy — segments with high overlap need careful ordering to avoid misclassification.

```python
segment_flags = ["CARDIO_METABOLIC", "MULTIMORBID_FRAIL", "MIDLIFE_ESCALATION", "ACUTE_OVERLAY"]

scored["segment_count"] = scored[segment_flags].sum(axis=1)

print("── Segment Count Distribution ───────────────────────")
print(scored["segment_count"].value_counts().sort_index())
print(f"\nPatients qualifying for 0 segments : {(scored['segment_count'] == 0).sum()}")
print(f"Patients qualifying for 1 segment  : {(scored['segment_count'] == 1).sum()}")
print(f"Patients qualifying for 2 segments : {(scored['segment_count'] == 2).sum()}")
print(f"Patients qualifying for 3+ segments: {(scored['segment_count'] >= 3).sum()}")

print("\n── Pairwise Overlap ─────────────────────────────────")
for i, seg1 in enumerate(segment_flags):
    for seg2 in segment_flags[i+1:]:
        overlap = ((scored[seg1] == 1) & (scored[seg2] == 1)).sum()
        pct = overlap / len(scored) * 100
        print(f"{seg1} + {seg2} : {overlap} patients ({pct:.1f}%)")
```

## Step 8 — Segment Assignment

The patient segmentation is based on clinical priority and evidence-based risk factors, ensuring mutually exclusive segments:

1. **Cardio-metabolic (Priority 1)**  
   - Patients with both cardiovascular disease and diabetes complications.  
   - Life-threatening events such as myocardial infarction are captured here.  
   **Rationale:** Cardiometabolic conditions are the most predictive of serious adverse outcomes and require the highest clinical attention.

2. **Multimorbid Frail (Priority 2)**  
   - Patients aged ≥66 with ≥3 chronic conditions **excluding Cardio-Metabolic**.  
   **Rationale:** Age and multimorbidity are strong predictors of adverse outcomes. Early management prevents escalation.

3. **Midlife Escalation (Priority 3)**  
   - Patients aged 41–65 with high acute utilization (≥50 encounters), **excluding higher-priority segments**.  
   **Rationale:** Early identification of high-utilization midlife patients can prevent progression to frailty; age-adjusted risk considered.

4. **Acute Overlay (Priority 4)**  
   - Patients with serious acute conditions (e.g., appendicitis, pneumonia, acute allergic reactions) not captured in higher-priority segments.  
   **Rationale:** Captures acute, potentially adverse events that require immediate attention. Life-threatening acute events like myocardial infarction are already assigned to the Cardio-Metabolic segment because of the hierarchy. Acute Overlay ensures that other serious acute conditions are not missed, allowing timely interventions.

5. **Low Risk Stable**  
   - Patients not qualifying for any of the above segments.  
   **Rationale:** Low predicted risk; standard care applies.

```python
SEGMENT_PRIORITY = [
    "CARDIO_METABOLIC",
    "MULTIMORBID_FRAIL",
    "MIDLIFE_ESCALATION",
    "ACUTE_OVERLAY",
]

SEGMENT_LABELS = {
    "CARDIO_METABOLIC":    "Cardio-Metabolic",
    "MULTIMORBID_FRAIL":   "Multimorbid Frail",
    "MIDLIFE_ESCALATION":  "Midlife Escalation",
    "ACUTE_OVERLAY":       "Acute Overlay",
}

def get_primary_segment(row):
    for seg in SEGMENT_PRIORITY:
        if row.get(seg, 0) == 1:
            return SEGMENT_LABELS[seg]
    return "Low Risk Stable"

def get_all_segments(row):
    qualifying = [
        SEGMENT_LABELS[seg]
        for seg in SEGMENT_PRIORITY
        if row.get(seg, 0) == 1
    ]
    return ", ".join(qualifying) if qualifying else "Low Risk Stable"

scored["primary_segment"] = scored.apply(get_primary_segment, axis=1)
scored["all_segments"]    = scored.apply(get_all_segments, axis=1)

print("── Primary Segment Distribution ─────────────────────")
print(scored["primary_segment"].value_counts())

print("\n── Segment Count Distribution ────────────────────────")
print(scored["segment_count"].value_counts().sort_index())

print(f"\nPatients with 2+ actionable segments : {(scored['segment_count'] >= 2).sum()}")
print(f"As % of cohort                       : {(scored['segment_count'] >= 2).mean()*100:.1f}%")

print("\n── Sample — all_segments column ──────────────────────")
print(scored[["PATIENT", "primary_segment", "all_segments"]].head(10))
```

## Segmentation Module Export

```python
%%writefile src/assign_segments.py
import pandas as pd
import os

SEGMENT_PRIORITY = [
    "CARDIO_METABOLIC",
    "MULTIMORBID_FRAIL",
    "MIDLIFE_ESCALATION",
    "ACUTE_OVERLAY",
]

SEGMENT_LABELS = {
    "CARDIO_METABOLIC":    "Cardio-Metabolic",
    "MULTIMORBID_FRAIL":   "Multimorbid Frail",
    "MIDLIFE_ESCALATION":  "Midlife Escalation",
    "ACUTE_OVERLAY":       "Acute Overlay",
}

ACUTE_OVERLAY_THRESHOLD      = 5
MULTIMORBID_FRAIL_AGE        = 66
MULTIMORBID_FRAIL_CATEGORIES = 3
MIDLIFE_ESCALATION_ENCOUNTERS = 50


def load_universe(universe_path):
    universe = pd.read_csv(universe_path)
    acute_codes = set(universe[universe["is_acute"] == 1]["CODE"])
    print(f"[segmentation] Acute SNOMED codes loaded : {len(acute_codes)}")
    return acute_codes


def flag_acute_overlay(scored, conditions, acute_codes):
    acute_condition_counts = (
        conditions[conditions["CODE"].isin(acute_codes)]
        .groupby("PATIENT")["CODE"]
        .nunique()
        .reset_index(name="acute_condition_count")
    )
    scored = scored.merge(acute_condition_counts, on="PATIENT", how="left")
    scored["acute_condition_count"] = scored["acute_condition_count"].fillna(0)
    scored["ACUTE_OVERLAY"] = (
        scored["acute_condition_count"] >= ACUTE_OVERLAY_THRESHOLD
    ).astype(int)
    print(f"[segmentation] Acute Overlay eligible    : {scored['ACUTE_OVERLAY'].sum()} ({scored['ACUTE_OVERLAY'].mean()*100:.1f}%)")
    return scored


def flag_cardio_metabolic(scored):
    scored["CARDIO_METABOLIC"] = (
        (scored["CARDIOVASCULAR"] == 1) &
        (scored["DIABETES & COMPLICATIONS"] == 1)
    ).astype(int)
    print(f"[segmentation] Cardio-Metabolic eligible : {scored['CARDIO_METABOLIC'].sum()} ({scored['CARDIO_METABOLIC'].mean()*100:.1f}%)")
    return scored


def flag_multimorbid_frail(scored):
    scored["MULTIMORBID_FRAIL"] = (
        (scored["age"] >= MULTIMORBID_FRAIL_AGE) &
        (scored["category_count"] >= MULTIMORBID_FRAIL_CATEGORIES) &
        (scored["CARDIO_METABOLIC"] == 0)
    ).astype(int)
    print(f"[segmentation] Multimorbid Frail eligible : {scored['MULTIMORBID_FRAIL'].sum()} ({scored['MULTIMORBID_FRAIL'].mean()*100:.1f}%)")
    return scored


def flag_midlife_escalation(scored):
    scored["MIDLIFE_ESCALATION"] = (
        (scored["age_band"] == "41-65") &
        (scored["encounter_count"] > MIDLIFE_ESCALATION_ENCOUNTERS)
    ).astype(int)
    print(f"[segmentation] Midlife Escalation eligible: {scored['MIDLIFE_ESCALATION'].sum()} ({scored['MIDLIFE_ESCALATION'].mean()*100:.1f}%)")
    return scored


def assign_segments(scored):
    def get_primary_segment(row):
        for seg in SEGMENT_PRIORITY:
            if row.get(seg, 0) == 1:
                return SEGMENT_LABELS[seg]
        return "Low Risk Stable"

    def get_all_segments(row):
        qualifying = [
            SEGMENT_LABELS[seg]
            for seg in SEGMENT_PRIORITY
            if row.get(seg, 0) == 1
        ]
        return ", ".join(qualifying) if qualifying else "Low Risk Stable"

    scored["primary_segment"] = scored.apply(get_primary_segment, axis=1)
    scored["all_segments"]    = scored.apply(get_all_segments, axis=1)
    return scored


def validate(scored):
    assert scored["primary_segment"].isna().sum() == 0, "WARNING: unassigned patients found"
    assert len(scored) == 1152, "WARNING: patient count mismatch"
    print(f"[segmentation] Validation PASSED — {len(scored)} patients, 0 unassigned")


def run(scored, conditions, universe_path, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)

    acute_codes = load_universe(universe_path)
    scored      = flag_cardio_metabolic(scored)
    scored      = flag_multimorbid_frail(scored)
    scored      = flag_midlife_escalation(scored)
    scored      = flag_acute_overlay(scored, conditions, acute_codes)
    scored      = assign_segments(scored)

    validate(scored)

    segment_summary = (
        scored.groupby("primary_segment")
        .agg(patient_count=("primary_segment", "count"))
        .reset_index()
    )
    segment_summary["percent"] = (
        100 * segment_summary["patient_count"] / len(scored)
    ).round(1)

    segments_path = os.path.join(output_dir, "patient_segments.csv")
    summary_path  = os.path.join(output_dir, "segment_summary.csv")
    scored.to_csv(segments_path, index=False)
    segment_summary.to_csv(summary_path, index=False)

    print(f"\n[segmentation] Exported : {segments_path}")
    print(f"[segmentation] Exported : {summary_path}")
    print()
    print("── Segment Summary ──────────────────────────────────")
    for _, row in segment_summary.iterrows():
        print(f"  {row['primary_segment']:<22} : {row['patient_count']:>4} patients ({row['percent']}%)")

    return scored, segment_summary
```

## Segment Distribution by Gender

Bar chart showing the percentage of patients in each primary segment, split by gender, with male (♂) and female (♀) symbols overlaid for visual clarity.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

gender_map = pd.read_csv("data/patients.csv", usecols=["Id", "GENDER"])
scored_tmp = scored.merge(gender_map, left_on="PATIENT", right_on="Id", how="left")

segment_summary = (
    scored_tmp.groupby(["primary_segment", "GENDER"])
    .size()
    .reset_index(name="count")
)

cohort_size = len(scored_tmp)
segment_summary["percent"] = segment_summary["count"] / cohort_size * 100

order = (
    segment_summary.groupby("primary_segment")["percent"]
    .sum()
    .sort_values(ascending=False)
    .index
)

palette_colors = {"M": "#1f77b4", "F": "#ff7f0e"}
symbols        = {"M": "♂", "F": "♀"}

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    y="percent",
    x="primary_segment",
    data=segment_summary,
    order=order,
    hue="GENDER",
    dodge=True,
    palette=palette_colors
)

# Build lookup: height → gender from actual data
height_gender_lookup = {}
for _, row in segment_summary.iterrows():
    h = round(row["percent"], 6)
    height_gender_lookup[h] = row["GENDER"]

SYMBOL_OFFSET = 0.8

for bar in ax.patches:
    height = bar.get_height()
    if height == 0:
        continue
    gender = height_gender_lookup.get(round(height, 6), None)
    if gender is None:
        continue
    x = bar.get_x() + bar.get_width() / 2
    ax.text(
        x=x,
        y=height + SYMBOL_OFFSET,
        s=symbols[gender],
        ha="center",
        va="bottom",
        fontsize=13,
        color="black"
    )

plt.ylabel("Percent of Cohort")
plt.xlabel("Segment")
plt.title("Segment Distribution by Gender")
plt.xticks(rotation=45, ha="right")
plt.ylim(0, segment_summary["percent"].max() + 8)
plt.tight_layout()
plt.savefig("outputs/figures/segment_gender_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Pairwise Segment Overlaps

Highlights patients belonging to overlapping segments, showing counts and percentages for key segment combinations. Only meaningful overlaps are included; other overlaps with very low counts (<1%) are excluded to avoid clutter and focus on actionable insights for cohort management.
 - Multimorbid Frail + Acute Overlay:
Patients with multiple chronic conditions and frailty are at high risk of acute complications, such as infections or decompensation, due to weakened immunity and limited physiological reserve. They require close monitoring and potentially urgent interventions.
- Midlife Escalation + Acute Overlay: 
Patients aged 41–65 who are beginning to develop cardiometabolic risks, early chronic conditions, or lifestyle-related vulnerabilities fall into this category. These individuals may experience periodic acute events—such as infections, flares of emerging chronic disease, or hospitalisations—that, if not addressed, create a chronic–acute feedback loop, where acute episodes worsen underlying conditions and accelerate progression toward multimorbidity. Early identification of these patients allows for preventive care, lifestyle interventions, and timely treatment, breaking the vicious cycle and slowing or halting the escalation from early chronic risk to complex, multimorbid states.
- Multimorbid Frail patients with concurrent acute conditions score an average of 19.0 — 
placing them firmly in the High tier. Midlife Escalation patients with acute overlay 
score 15.9 on average, reflecting earlier-stage complexity where preventative intervention 
remains viable. The 3.1 point gap between groups validates the segment priority ordering.

## Patient Distribution by Risk Tier within Segments

Shows a horizontal 100% stacked bar chart of risk tiers across each primary segment. Each bar represents a segment, with colour-coded sections for Low, Medium, High, and Critical tiers, allowing easy comparison of risk composition between segments.

```python
import pandas as pd
import matplotlib.pyplot as plt

patients_df = pd.read_csv("outputs/patient_segments.csv")

agg_df = patients_df.groupby(["primary_segment", "risk_tier"]).size().reset_index(name="patient_count")

pivot_df = agg_df.pivot(index="primary_segment", columns="risk_tier", values="patient_count").fillna(0)
risk_order = ["Low", "Medium", "High", "Critical"]
pivot_df = pivot_df[risk_order]

pivot_perc = pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100

segment_order = ["Cardio-Metabolic", "Multimorbid Frail", "Midlife Escalation", "Acute Overlay", "Low Risk Stable"]
pivot_perc = pivot_perc.reindex(segment_order)

fig, ax = plt.subplots(figsize=(10, 6))

colors = ["#6baed6", "#a8d5a2", "#FFC107", "#e6550d"]

pivot_perc.plot(kind="barh", stacked=True, color=colors, ax=ax)

for i, segment in enumerate(pivot_perc.index):
    left = 0
    for j, tier in enumerate(risk_order):
        val = pivot_perc.loc[segment, tier]
        if val > 5:
            ax.text(
                left + val / 2, i,
                f"{val:.0f}%",
                ha="center", va="center",
                color="white", fontsize=10, fontweight="bold"
            )
        left += val

ax.set_xlabel("Percentage of Patients")
ax.set_ylabel("Primary Segment")
ax.set_title("Patient Distribution by Risk Tier within Segment")
ax.legend(title="Risk Tier", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig("outputs/figures/patient_distribution_primary_segment.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Step 2 — SNOMED to ICD-10 Mapping Table
Builds an explicit one-to-one mapping table between SNOMED CT codes and UK ICD-10 codes. 
Each SNOMED code is assigned its own row with a clinically specific ICD-10 description, 
preserving the diagnostic nuance of the original SNOMED concept even where multiple SNOMED 
codes share the same ICD-10 code.

**Why this matters:** A simple dictionary mapping SNOMED → ICD-10 collapses clinical 
detail — for example, diabetic nephropathy and microalbuminuria both map to E11.2, but 
represent different stages of renal involvement. This table preserves that distinction 
at the SNOMED level whilst producing valid ICD-10 output for reporting. This mirrors 
how NHS clinical coding systems handle terminology translation — SNOMED drives clinical 
logic, ICD-10 drives administrative reporting.

The mapping table is saved in two formats — `data/reference/snomed_to_icd10_map.csv` 
for human review and audit, and `data/reference/snomed_to_icd10_map.json` for 
programmatic use in the FHIR export pipeline.

For this MVP a curated crosswalk is used. For enterprise deployment this module is 
designed to integrate with the NHS TRUD API for live SNOMED-to-ICD-10 mapping, 
replacing the static crosswalk with a dynamically validated terminology service.

## ICD10 Mapping Module Export

```python
%%writefile src/map_icd10.py
import pandas as pd
import json
import os

FALLBACK_ICD10_CODE = "Z99.9"
FALLBACK_ICD10_DESC = "Unknown SNOMED code - not in mapping table"
HYPERTENSION_SNOMED_CODE = 59621000
HYPERTENSION_CORRECT_DESC = "Essential hypertension (disorder)"


def load_mapping(map_path):
    df_map = pd.read_csv(map_path)
    print(f"[icd10_mapper] Mapping rows loaded       : {len(df_map)}")
    return df_map


def enrich_universe(universe, df_map, universe_path):
    for col in ["icd10_code", "icd10_desc"]:
        if col in universe.columns:
            universe = universe.drop(columns=[col])

    universe = universe.merge(
        df_map[["snomed_code", "icd10_code", "icd10_desc"]].rename(columns={"snomed_code": "CODE"}),
        on="CODE",
        how="left"
    )

    universe.loc[universe["CODE"] == HYPERTENSION_SNOMED_CODE, "DESCRIPTION"] = HYPERTENSION_CORRECT_DESC

    universe.to_csv(universe_path, index=False)
    print(f"[icd10_mapper] Universe enriched         : {len(universe)} codes")
    print(f"[icd10_mapper] Codes with ICD-10         : {universe['icd10_code'].notna().sum()}")
    print(f"[icd10_mapper] Codes without ICD-10      : {universe['icd10_code'].isna().sum()}")
    return universe


def map_conditions(conditions, universe):
    conditions_mapped = conditions.merge(
        universe[["CODE", "DESCRIPTION", "CATEGORY", "icd10_code", "icd10_desc"]],
        on=["CODE", "DESCRIPTION"],
        how="left"
    )

    unmapped = conditions_mapped["icd10_code"].isna().sum()
    if unmapped > 0:
        print(f"[icd10_mapper] WARNING: {unmapped} rows unmapped — applying {FALLBACK_ICD10_CODE} fallback")
        conditions_mapped["icd10_code"] = conditions_mapped["icd10_code"].fillna(FALLBACK_ICD10_CODE)
        conditions_mapped["icd10_desc"] = conditions_mapped["icd10_desc"].fillna(FALLBACK_ICD10_DESC)
    else:
        print(f"[icd10_mapper] Unmapped fallback check   : PASSED — all codes mapped")

    print(f"[icd10_mapper] Condition rows mapped     : {len(conditions_mapped)}")
    return conditions_mapped


def build_patient_summary(conditions_mapped, scored):
    patient_icd10_summary = (
        conditions_mapped.groupby("PATIENT")["icd10_code"]
        .apply(lambda x: sorted(x.dropna().unique().tolist()))
        .reset_index()
        .rename(columns={"icd10_code": "icd10_codes_all"})
    )

    scored_with_icd10 = scored.merge(patient_icd10_summary, on="PATIENT", how="left")
    print(f"[icd10_mapper] Patients with ICD-10 summary : {len(scored_with_icd10)}")
    return scored_with_icd10


def run(conditions, universe, universe_path, map_path, scored, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)
    os.makedirs("data/reference", exist_ok=True)

    df_map = load_mapping(map_path)
    universe = enrich_universe(universe, df_map, universe_path)
    conditions_mapped = map_conditions(conditions, universe)
    scored_with_icd10 = build_patient_summary(conditions_mapped, scored)

    conditions_path = os.path.join(output_dir, "conditions_icd10_mapped.csv")
    conditions_mapped.to_csv(conditions_path, index=False)

    scored_export = scored_with_icd10.copy()
    scored_export["icd10_codes_all"] = scored_export["icd10_codes_all"].apply(
        lambda x: ", ".join(x) if isinstance(x, list) else x
    )
    scores_path = os.path.join(output_dir, "patient_risk_scores.csv")
    scored_export.to_csv(scores_path, index=False)

    print(f"\n[icd10_mapper] Exported : {conditions_path}")
    print(f"[icd10_mapper] Exported : {scores_path}")

    return conditions_mapped, scored_with_icd10, universe
```

## Step 1 — Clean the SNOMED–ICD10 Data

This step ensures the dataset is ready for encounter-level merging:

1. **Load the raw mapped file** (`conditions_icd10_mapped.csv`)  
2. **Check for required columns**: `PATIENT`, `ENCOUNTER`, `START`, `STOP`, `CODE`, `DESCRIPTION`, `CATEGORY`, `icd10_code`, `icd10_desc`  
3. **Remove exact duplicate rows** (same patient, encounter, SNOMED code, ICD-10 mapping)  
4. **Save a cleaned file** (`conditions_icd10_cleaned.csv`) for downstream processing  

This guarantees that the dataset is deduplicated and all necessary fields are present for FHIR-ready merging.

```python
import pandas as pd
import os

# --- Load mapped SNOMED-ICD10 file ---
file_path = "outputs/conditions_icd10_mapped.csv"
df = pd.read_csv(file_path, parse_dates=["START", "STOP"])

# --- Inspect required columns ---
required_cols = ["PATIENT", "ENCOUNTER", "START", "STOP", "CODE", "DESCRIPTION", "CATEGORY", "icd10_code", "icd10_desc"]
missing_cols = [col for col in required_cols if col not in df.columns]
if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")
print("All required columns present.")

# --- Remove exact duplicates ---
before_dedup = len(df)
df.drop_duplicates(inplace=True)
print(f"Dropped {before_dedup - len(df)} exact duplicate rows.")

# --- Fill missing values in critical columns ---
df["CODE"] = df["CODE"].astype(str)
df["DESCRIPTION"] = df["DESCRIPTION"].fillna("Unknown")
df["CATEGORY"] = df["CATEGORY"].fillna("Uncategorized")
df["icd10_code"] = df["icd10_code"].fillna("Unknown")
df["icd10_desc"] = df["icd10_desc"].fillna("Unknown")

# --- Ensure START and STOP are datetime ---
df["START"] = pd.to_datetime(df["START"], errors="coerce")
df["STOP"] = pd.to_datetime(df["STOP"], errors="coerce")

# --- Optional: remove rows with invalid START dates ---
invalid_start = df["START"].isna().sum()
if invalid_start > 0:
    df = df.dropna(subset=["START"])
    print(f"Dropped {invalid_start} rows with invalid START dates.")

# --- Save cleaned intermediate file ---
os.makedirs("outputs", exist_ok=True)
clean_path = "outputs/conditions_icd10_cleaned.csv"
df.to_csv(clean_path, index=False)
print(f"Cleaned file saved: {len(df)} rows")
```

## Step 2 — Encounter-Level Merge
Groups condition rows by patient and encounter, merging multiple SNOMED codes recorded 
at the same visit into a single row. This reflects the clinical reality that one encounter 
can produce multiple diagnoses — in FHIR R4 these map to one Encounter resource with 
multiple linked Condition resources. Grouping is performed on PATIENT, ENCOUNTER, and 
START date. STOP date is excluded from the groupby key to prevent NaT values causing 
incorrect row splits for ongoing chronic conditions. Manual review confirms all 
multi-SNOMED merges are clinically valid co-diagnoses recorded at a single visit.

```python
conditions_mapped = pd.read_csv("outputs/conditions_icd10_cleaned.csv", parse_dates=["START", "STOP"])

encounter_merged = (
    conditions_mapped
    .groupby(["PATIENT", "ENCOUNTER", "START"], dropna=False)
    .agg(
        snomed_codes  = ("CODE",          lambda x: sorted(x.unique().tolist())),
        snomed_count  = ("CODE",          "nunique"),
        icd10_codes   = ("icd10_code",    lambda x: sorted(x.dropna().unique().tolist())),
        descriptions  = ("DESCRIPTION",   lambda x: sorted(x.unique().tolist())),
        stop_date     = ("STOP",          "first"),
    )
    .reset_index()
    .rename(columns={"stop_date": "STOP"})
)

encounter_merged.to_csv("outputs/conditions_encounter_merged.csv", index=False)

print(f"Total encounter rows          : {len(encounter_merged)}")
print(f"Encounters with 1 SNOMED      : {(encounter_merged['snomed_count'] == 1).sum()}")
print(f"Encounters with 2+ SNOMED     : {(encounter_merged['snomed_count'] >= 2).sum()}")
print(f"Max SNOMED codes per encounter : {encounter_merged['snomed_count'].max()}")
```

clinical terminology throughout — ICD-10 is carried as a derived administrative code only.

```python
import json
import os
import pandas as pd
from datetime import datetime, timezone

# ── Load required data ────────────────────────────────────────────
encounter_merged = pd.read_csv("outputs/conditions_encounter_merged.csv", parse_dates=["START", "STOP"])
patients         = pd.read_csv("data/patients.csv")
scored_segments  = pd.read_csv("outputs/patient_segments.csv")

# ── Display correction map ────────────────────────────────────────
DISPLAY_CORRECTIONS = {
    "N39.0":  "Urinary tract infection, site not specified",
    "E11.9":  "Type 2 diabetes mellitus : Without complications",
    "E11.2":  "Type 2 diabetes mellitus : With renal complications",
    "E11.3":  "Type 2 diabetes mellitus : With ophthalmic complications",
    "E11.4":  "Type 2 diabetes mellitus : With neurological complications",
    "O03.9":  "Spontaneous abortion : Complete or unspecified, without complication",
    "I48.9":  "Atrial fibrillation and atrial flutter, unspecified",
    "I46.9":  "Cardiac arrest, unspecified",
    "R73.0":  "Abnormal glucose tolerance test",
    "G30.9":  "Alzheimer disease, unspecified",
    "G30.0":  "Alzheimer disease with early onset",
    "C50.9":  "Breast, unspecified",
    "S62.0":  "Fracture of navicular [scaphoid] bone of hand",
    "S63.6":  "Sprain and strain of finger(s)",
    "S93.4":  "Sprain and strain of ankle",
    "S91.3":  "Open wound of other parts of foot",
    "T50.9":  "Other and unspecified drugs, medicaments and biological substances",
    "O02.0":  "Blighted ovum and nonhydatidiform mole",
    "O36.9":  "Maternal care for fetal problem, unspecified",
}

# ── Helper functions ──────────────────────────────────────────────
def safe_date(dt):
    if pd.isna(dt):
        return None
    return pd.Timestamp(dt).strftime("%Y-%m-%d")

def parse_list_col(val):
    if isinstance(val, list):
        return val
    try:
        return json.loads(val.replace("'", '"'))
    except Exception:
        return []

def strip_display(desc):
    if " - " in str(desc):
        return desc.split(" - ")[0].strip()
    return desc

def get_official_display(icd10_code, current_display):
    return DISPLAY_CORRECTIONS.get(icd10_code, current_display)

def build_patient_resource(row):
    gender_map = {"M": "male", "F": "female", "m": "male", "f": "female"}
    gender     = gender_map.get(str(row["GENDER"]).strip(), "unknown") if pd.notna(row["GENDER"]) else "unknown"
    return {
        "resourceType": "Patient",
        "id":           row["Id"],
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Patient {row['Id']} | Gender: {gender} | DOB: {safe_date(pd.to_datetime(row['BIRTHDATE'], errors='coerce'))}</div>"
        },
        "gender":    gender,
        "birthDate": safe_date(pd.to_datetime(row["BIRTHDATE"], errors="coerce")),
    }

def build_encounter_resource(enc_row):
    enc_id = f"{enc_row['ENCOUNTER']}-{pd.Timestamp(enc_row['START']).strftime('%Y%m%d')}"
    period = {"start": safe_date(enc_row["START"])}
    if pd.notna(enc_row["STOP"]):
        period["end"] = safe_date(enc_row["STOP"])
    return {
        "resourceType": "Encounter",
        "id":           enc_id,
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Encounter on {safe_date(enc_row['START'])} for patient {enc_row['PATIENT']}</div>"
        },
        "status": "finished",
        "class": {
            "system":  "http://terminology.hl7.org/CodeSystem/v3-ActCode",
            "code":    "AMB",
            "display": "ambulatory"
        },
        "subject": {"reference": f"Patient/{enc_row['PATIENT']}"},
        "period":  period
    }

def build_condition_resources(enc_row, df_map):
    snomed_codes = parse_list_col(enc_row["snomed_codes"])
    icd10_codes  = parse_list_col(enc_row["icd10_codes"])
    descriptions = parse_list_col(enc_row["descriptions"])
    conditions   = []
    enc_id       = f"{enc_row['ENCOUNTER']}-{pd.Timestamp(enc_row['START']).strftime('%Y%m%d')}"

    for i, snomed_code in enumerate(snomed_codes):
        snomed_str = str(snomed_code)
        match      = df_map[df_map["snomed_code"].astype(str) == snomed_str]

        if not match.empty:
            icd10_code  = match.iloc[0]["icd10_code"]
            icd10_desc  = get_official_display(icd10_code, strip_display(match.iloc[0]["icd10_desc"]))
            snomed_desc = match.iloc[0]["DESCRIPTION"] if "DESCRIPTION" in match.columns else (descriptions[i] if i < len(descriptions) else "Unknown")
        else:
            icd10_code  = icd10_codes[i] if i < len(icd10_codes) else "Unknown"
            icd10_desc  = get_official_display(icd10_code, "Unknown")
            snomed_desc = descriptions[i] if i < len(descriptions) else "Unknown"

        cond_id   = f"{enc_id}-{i}"
        condition = {
            "resourceType": "Condition",
            "id":           cond_id,
            "text": {
                "status": "generated",
                "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>{snomed_desc} (SNOMED: {snomed_str} | ICD-10: {icd10_code})</div>"
            },
            "subject":       {"reference": f"Patient/{enc_row['PATIENT']}"},
            "encounter":     {"reference": f"Encounter/{enc_id}"},
            "onsetDateTime": safe_date(enc_row["START"]),
            "code": {
                "coding": [
                    {
                        "system":       "http://snomed.info/sct",
                        "code":         snomed_str,
                        "display":      snomed_desc,
                        "userSelected": True
                    },
                    {
                        "system":       "http://hl7.org/fhir/sid/icd-10",
                        "code":         icd10_code,
                        "display":      icd10_desc,
                        "userSelected": False
                    }
                ],
                "text": snomed_desc
            },
            "clinicalStatus": {
                "coding": [{
                    "system":  "http://terminology.hl7.org/CodeSystem/condition-clinical",
                    "code":    "active" if pd.isna(enc_row["STOP"]) else "resolved",
                    "display": "Active" if pd.isna(enc_row["STOP"]) else "Resolved"
                }]
            }
        }
        conditions.append(condition)
    return conditions

def build_risk_assessment(patient_id, scored_row):
    return {
        "resourceType": "RiskAssessment",
        "id":           patient_id,
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Risk tier: {scored_row['risk_tier']} | Score: {scored_row['risk_score']} | Segment: {scored_row['primary_segment']}</div>"
        },
        "subject": {"reference": f"Patient/{patient_id}"},
        "status":  "final",
        "prediction": [
            {
                "outcome": {
                    "coding": [{
                        "system":  "http://snomed.info/sct",
                        "code":    "225338004",
                        "display": "Risk assessment"
                    }]
                },
                "probabilityDecimal": float(scored_row["risk_score"]),
            }
        ],
        "note": [
            {
                "text": (
                    f"Risk tier: {scored_row['risk_tier']} | "
                    f"Primary segment: {scored_row['primary_segment']} | "
                    f"All segments: {scored_row['all_segments']}"
                )
            }
        ]
    }

# ── Load SNOMED → ICD-10 mapping table ───────────────────────────
df_map = pd.read_csv("data/reference/snomed_to_icd10_map.csv")
df_map["icd10_desc"] = df_map["icd10_desc"].str.replace("—", "-", regex=False)

universe_ref = pd.read_csv("outputs/snomed_conditions_universe.csv")
universe_ref.loc[universe_ref["CODE"] == 59621000,  "DESCRIPTION"] = "Essential hypertension (disorder)"
universe_ref.loc[universe_ref["CODE"] == 403190006, "DESCRIPTION"] = "Epidermal burn of skin (disorder)"
universe_ref.loc[universe_ref["CODE"] == 5602001,   "DESCRIPTION"] = "Harmful pattern of use of opioid"
universe_ref.loc[universe_ref["CODE"] == 201834006, "DESCRIPTION"] = "Localized, primary osteoarthritis of the hand"

df_map = df_map.merge(
    universe_ref[["CODE", "DESCRIPTION"]].rename(columns={"CODE": "snomed_code"}),
    on="snomed_code",
    how="left"
)

# ── Build FHIR Bundle ─────────────────────────────────────────────
bundle = {
    "resourceType": "Bundle",
    "id":           f"clinical-risk-bundle-{datetime.today().strftime('%Y-%m-%d')}",
    "type":         "collection",
    "timestamp":    datetime.now(timezone.utc).strftime("%Y-%m-%dT%H:%M:%S+00:00"),
    "entry":        []
}

scored_index   = scored_segments.set_index("PATIENT")
enc_by_patient = encounter_merged.groupby("PATIENT")

patient_count   = 0
condition_count = 0
encounter_count = 0
risk_count      = 0

for _, pat_row in patients.iterrows():
    patient_id = pat_row["Id"]

    if patient_id not in scored_index.index:
        continue

    scored_row = scored_index.loc[patient_id]

    bundle["entry"].append({
        "fullUrl":  f"http://clinical-risk-pipeline/Patient/{pat_row['Id']}",
        "resource": build_patient_resource(pat_row)
    })
    patient_count += 1

    if patient_id in enc_by_patient.groups:
        for _, enc_row in enc_by_patient.get_group(patient_id).iterrows():
            enc_resource = build_encounter_resource(enc_row)
            bundle["entry"].append({
                "fullUrl":  f"http://clinical-risk-pipeline/Encounter/{enc_resource['id']}",
                "resource": enc_resource
            })
            encounter_count += 1

            for cond in build_condition_resources(enc_row, df_map):
                bundle["entry"].append({
                    "fullUrl":  f"http://clinical-risk-pipeline/Condition/{cond['id']}",
                    "resource": cond
                })
                condition_count += 1

    bundle["entry"].append({
        "fullUrl":  f"http://clinical-risk-pipeline/RiskAssessment/{patient_id}",
        "resource": build_risk_assessment(patient_id, scored_row)
    })
    risk_count += 1

# ── Export ────────────────────────────────────────────────────────
os.makedirs("outputs/fhir", exist_ok=True)
output_path = "outputs/fhir/clinical_risk_bundle.json"

with open(output_path, "w") as f:
    json.dump(bundle, f, indent=2, default=str, ensure_ascii=False)

print(f"── FHIR Bundle Export ───────────────────────────────")
print(f"Output file        : {output_path}")
print(f"Total entries      : {len(bundle['entry'])}")
print(f"  Patient          : {patient_count}")
print(f"  Encounter        : {encounter_count}")
print(f"  Condition        : {condition_count}")
print(f"  RiskAssessment   : {risk_count}")
print(f"File size          : {os.path.getsize(output_path) / 1024:.1f} KB")
```

## Step 4 — Structural Validation
Validates each resource in the bundle against required FHIR R4 fields. Checks cover 
required fields per resource type, date format conformance to ISO 8601, subject 
reference integrity, and fullUrl uniqueness across all 18,337 entries.

```python
import re
import json

with open("outputs/fhir/clinical_risk_bundle.json") as f:
    bundle = json.load(f)

errors       = []
counts       = {"Patient": 0, "Encounter": 0, "Condition": 0, "RiskAssessment": 0}
date_pattern = re.compile(r"^\d{4}-\d{2}-\d{2}$")
date_errors  = []

for i, entry in enumerate(bundle.get("entry", [])):
    res   = entry.get("resource", {})
    rtype = res.get("resourceType")

    if not rtype:
        errors.append(f"Entry {i}: Missing resourceType")
        continue

    if rtype in counts:
        counts[rtype] += 1

    if rtype == "Patient":
        if "id" not in res:
            errors.append(f"Entry {i} Patient: missing id")
        if "gender" not in res:
            errors.append(f"Entry {i} Patient: missing gender")
        if "birthDate" not in res:
            errors.append(f"Entry {i} Patient: missing birthDate")
        bd = res.get("birthDate", "")
        if bd and not date_pattern.match(str(bd)):
            date_errors.append(f"Entry {i} Patient: invalid birthDate format {bd}")

    elif rtype == "Encounter":
        if "subject" not in res:
            errors.append(f"Entry {i} Encounter: missing subject")
        if "period" not in res:
            errors.append(f"Entry {i} Encounter: missing period")
        if "status" not in res:
            errors.append(f"Entry {i} Encounter: missing status")
        if "class" not in res:
            errors.append(f"Entry {i} Encounter: missing class (required in FHIR R4)")
        for field in ["start", "end"]:
            val = res.get("period", {}).get(field, "")
            if val and not date_pattern.match(str(val)):
                date_errors.append(f"Entry {i} Encounter: invalid {field} format {val}")

    elif rtype == "Condition":
        if "subject" not in res:
            errors.append(f"Entry {i} Condition: missing subject")
        if "code" not in res:
            errors.append(f"Entry {i} Condition: missing code")
        if "encounter" not in res:
            errors.append(f"Entry {i} Condition: missing encounter reference")
        if "clinicalStatus" not in res:
            errors.append(f"Entry {i} Condition: missing clinicalStatus")
        codings = res.get("code", {}).get("coding", [])
        systems = [c.get("system") for c in codings]
        if "http://snomed.info/sct" not in systems:
            errors.append(f"Entry {i} Condition: missing SNOMED CT coding")
        if "http://hl7.org/fhir/sid/icd-10" not in systems:
            errors.append(f"Entry {i} Condition: missing ICD-10 coding")
        val = res.get("onsetDateTime", "")
        if val and not date_pattern.match(str(val)):
            date_errors.append(f"Entry {i} Condition: invalid onsetDateTime format {val}")

    elif rtype == "RiskAssessment":
        if "subject" not in res:
            errors.append(f"Entry {i} RiskAssessment: missing subject")
        if "prediction" not in res:
            errors.append(f"Entry {i} RiskAssessment: missing prediction")
        if "status" not in res:
            errors.append(f"Entry {i} RiskAssessment: missing status")
        if "note" not in res:
            errors.append(f"Entry {i} RiskAssessment: missing note")

# Subject reference validation
patient_ids = set(
    e["resource"]["id"]
    for e in bundle["entry"]
    if e["resource"]["resourceType"] == "Patient"
)

reference_errors = []
for i, entry in enumerate(bundle["entry"]):
    res   = entry["resource"]
    rtype = res.get("resourceType")
    if rtype in ["Encounter", "Condition", "RiskAssessment"]:
        ref         = res.get("subject", {}).get("reference", "")
        patient_ref = ref.replace("Patient/", "")
        if patient_ref not in patient_ids:
            reference_errors.append(f"Entry {i} {rtype}: subject reference {ref} has no matching Patient")

# fullUrl validation
fullurl_errors = []
seen_urls = {}
for i, entry in enumerate(bundle["entry"]):
    url = entry.get("fullUrl", "")
    if not url:
        fullurl_errors.append(f"Entry {i}: missing fullUrl")
    elif url in seen_urls:
        fullurl_errors.append(f"Entry {i}: duplicate fullUrl {url} (first seen at entry {seen_urls[url]})")
    else:
        seen_urls[url] = i

# Print results
print("── Resource Counts ──────────────────────────────────")
for rtype, count in counts.items():
    print(f"  {rtype:<20} : {count}")
print(f"  {'Total entries':<20} : {len(bundle['entry'])}")

print("\n── Date Format Validation ───────────────────────────")
if date_errors:
    print(f"FAILED — {len(date_errors)} date format errors")
    for e in date_errors[:10]:
        print(f"  - {e}")
else:
    print("PASSED — all dates conform to ISO 8601 YYYY-MM-DD")

print("\n── Subject Reference Validation ─────────────────────")
if reference_errors:
    print(f"FAILED — {len(reference_errors)} reference errors")
    for e in reference_errors[:10]:
        print(f"  - {e}")
else:
    print("PASSED — all references match a Patient resource")

print("\n── fullUrl Validation ───────────────────────────────")
if fullurl_errors:
    print(f"FAILED — {len(fullurl_errors)} fullUrl errors")
    for e in fullurl_errors[:10]:
        print(f"  - {e}")
else:
    print("PASSED — all fullUrls present and unique")

print("\n── Structural Validation ────────────────────────────")
if errors:
    print(f"FAILED — {len(errors)} structural errors found:")
    for e in errors[:20]:
        print(f"  - {e}")
    if len(errors) > 20:
        print(f"  ... and {len(errors) - 20} more")
else:
    print("PASSED — all resources passed structural validation.")
```

## Step 5 — Comprehensive FHIR R4 Validation
Nine-point validation suite covering bundle-level integrity, entry completeness, 
per-resource field requirements, reference consistency, fullUrl uniqueness, and 
risk score sanity checks. Designed to catch structural issues that the HL7 FHIR 
Validator identified in the initial export — including null period end dates, 
missing Encounter class, and duplicate fullUrls.

```python
import json
import re
from collections import defaultdict

with open("outputs/fhir/clinical_risk_bundle.json") as f:
    bundle = json.load(f)

print("══ Comprehensive FHIR R4 Validation ══════════════════")

# ── 1. Bundle-level checks ────────────────────────────────
print("\n── 1. Bundle Level ──────────────────────────────────")
assert bundle.get("resourceType") == "Bundle",       "FAIL: resourceType must be Bundle"
assert bundle.get("type") == "collection",           "FAIL: Bundle type must be collection"
assert "entry" in bundle,                             "FAIL: Bundle missing entry array"
assert "timestamp" in bundle,                         "FAIL: Bundle missing timestamp"
print("  resourceType     : PASSED")
print("  type             : PASSED")
print("  entry present    : PASSED")
print("  timestamp present: PASSED")

# ── 2. Entry-level checks ─────────────────────────────────
print("\n── 2. Entry Level ───────────────────────────────────")
missing_resource  = sum(1 for e in bundle["entry"] if "resource" not in e)
missing_fullurl   = sum(1 for e in bundle["entry"] if "fullUrl" not in e)
print(f"  Entries missing resource : {missing_resource}")
print(f"  Entries missing fullUrl  : {missing_fullurl}")

# ── 3. Patient validation ─────────────────────────────────
print("\n── 3. Patient Resources ─────────────────────────────")
patients    = [e["resource"] for e in bundle["entry"] if e["resource"].get("resourceType") == "Patient"]
patient_ids = {p["id"] for p in patients}
valid_genders = {"male", "female", "other", "unknown"}
gender_errors = [p["id"] for p in patients if p.get("gender") not in valid_genders]
missing_birth = [p["id"] for p in patients if "birthDate" not in p]
print(f"  Total patients           : {len(patients)}")
print(f"  Invalid gender values    : {len(gender_errors)}")
print(f"  Missing birthDate        : {len(missing_birth)}")

# ── 4. Encounter validation ───────────────────────────────
print("\n── 4. Encounter Resources ───────────────────────────")
encounters       = [e["resource"] for e in bundle["entry"] if e["resource"].get("resourceType") == "Encounter"]
missing_class    = [e["id"] for e in encounters if "class" not in e]
missing_period   = [e["id"] for e in encounters if "period" not in e]
null_end_dates   = [e["id"] for e in encounters if e.get("period", {}).get("end") is None and "end" in e.get("period", {})]
orphan_enc       = [e["id"] for e in encounters if e.get("subject", {}).get("reference", "").replace("Patient/", "") not in patient_ids]
print(f"  Total encounters         : {len(encounters)}")
print(f"  Missing class field      : {len(missing_class)}")
print(f"  Missing period           : {len(missing_period)}")
print(f"  Null end dates           : {len(null_end_dates)}")
print(f"  Orphaned encounters      : {len(orphan_enc)}")

# ── 5. Condition validation ───────────────────────────────
print("\n── 5. Condition Resources ───────────────────────────")
conditions       = [e["resource"] for e in bundle["entry"] if e["resource"].get("resourceType") == "Condition"]
enc_ids          = {e["id"] for e in encounters}
missing_snomed   = []
missing_icd10    = []
missing_enc_ref  = []
orphan_cond      = []
invalid_status   = []
valid_statuses   = {"active", "recurrence", "relapse", "inactive", "remission", "resolved"}

for c in conditions:
    codings = c.get("code", {}).get("coding", [])
    systems = [x.get("system") for x in codings]
    if "http://snomed.info/sct" not in systems:
        missing_snomed.append(c["id"])
    if "http://hl7.org/fhir/sid/icd-10" not in systems:
        missing_icd10.append(c["id"])
    enc_ref = c.get("encounter", {}).get("reference", "").replace("Encounter/", "")
    if not enc_ref:
        missing_enc_ref.append(c["id"])
    elif enc_ref not in enc_ids:
        orphan_cond.append(c["id"])
    clinical_status = c.get("clinicalStatus", {}).get("coding", [{}])[0].get("code", "")
    if clinical_status not in valid_statuses:
        invalid_status.append(c["id"])
    pat_ref = c.get("subject", {}).get("reference", "").replace("Patient/", "")
    if pat_ref not in patient_ids:
        orphan_cond.append(c["id"])

print(f"  Total conditions         : {len(conditions)}")
print(f"  Missing SNOMED coding    : {len(missing_snomed)}")
print(f"  Missing ICD-10 coding    : {len(missing_icd10)}")
print(f"  Missing encounter ref    : {len(missing_enc_ref)}")
print(f"  Orphaned conditions      : {len(orphan_cond)}")
print(f"  Invalid clinicalStatus   : {len(invalid_status)}")

# ── 6. RiskAssessment validation ──────────────────────────
print("\n── 6. RiskAssessment Resources ──────────────────────")
risk_assessments   = [e["resource"] for e in bundle["entry"] if e["resource"].get("resourceType") == "RiskAssessment"]
missing_prediction = [r["id"] for r in risk_assessments if not r.get("prediction")]
invalid_score      = [r["id"] for r in risk_assessments if not isinstance(r.get("prediction", [{}])[0].get("probabilityDecimal"), (int, float))]
missing_note       = [r["id"] for r in risk_assessments if not r.get("note")]
orphan_risk        = [r["id"] for r in risk_assessments if r.get("subject", {}).get("reference", "").replace("Patient/", "") not in patient_ids]
print(f"  Total RiskAssessments    : {len(risk_assessments)}")
print(f"  Missing prediction       : {len(missing_prediction)}")
print(f"  Invalid score type       : {len(invalid_score)}")
print(f"  Missing note             : {len(missing_note)}")
print(f"  Orphaned assessments     : {len(orphan_risk)}")

# ── 7. Completeness checks ────────────────────────────────
print("\n── 7. Completeness ──────────────────────────────────")
patients_with_risk = {r.get("subject", {}).get("reference", "").replace("Patient/", "") for r in risk_assessments}
patients_missing_risk = patient_ids - patients_with_risk
patients_with_enc  = {e.get("subject", {}).get("reference", "").replace("Patient/", "") for e in encounters}
patients_missing_enc = patient_ids - patients_with_enc
print(f"  Patients missing RiskAssessment : {len(patients_missing_risk)}")
print(f"  Patients missing Encounter      : {len(patients_missing_enc)}")

# ── 8. Duplicate fullUrl check ────────────────────────────
print("\n── 8. fullUrl Uniqueness ────────────────────────────")
seen_urls    = {}
dup_urls     = []
for i, entry in enumerate(bundle["entry"]):
    url = entry.get("fullUrl", "")
    if url in seen_urls:
        dup_urls.append(f"Entry {i}: {url} (duplicate of entry {seen_urls[url]})")
    else:
        seen_urls[url] = i
print(f"  Duplicate fullUrls : {len(dup_urls)}")
if dup_urls:
    for d in dup_urls[:5]:
        print(f"    - {d}")

# ── 9. Score distribution sanity check ───────────────────
print("\n── 9. Risk Score Sanity ─────────────────────────────")
scores = [r.get("prediction", [{}])[0].get("probabilityDecimal", 0) for r in risk_assessments]
print(f"  Min score  : {min(scores):.2f}")
print(f"  Max score  : {max(scores):.2f}")
print(f"  Mean score : {sum(scores)/len(scores):.2f}")
print(f"  Zero scores: {sum(1 for s in scores if s == 0)}")

# ── 10. Final summary ─────────────────────────────────────
print("\n══ Validation Summary ════════════════════════════════")
all_clear = all([
    missing_resource == 0,
    missing_fullurl == 0,
    len(gender_errors) == 0,
    len(missing_birth) == 0,
    len(missing_class) == 0,
    len(null_end_dates) == 0,
    len(orphan_enc) == 0,
    len(missing_snomed) == 0,
    len(missing_icd10) == 0,
    len(orphan_cond) == 0,
    len(invalid_status) == 0,
    len(missing_prediction) == 0,
    len(invalid_score) == 0,
    len(orphan_risk) == 0,
    len(patients_missing_risk) == 0,
    len(dup_urls) == 0,
])
if all_clear:
    print("ALL CHECKS PASSED")
else:
    print("ISSUES FOUND — review sections above")
print(f"Total resources validated : {len(bundle['entry'])}")
```

## Step 6 — Cross-Validation Against Source Data
End-to-end verification that no data was lost or corrupted during the FHIR export 
pipeline. Compares patient counts, encounter counts, condition counts, RiskAssessment 
counts, SNOMED code coverage, patient ID consistency, risk scores, and segment 
assignments between the source analytical tables and the exported FHIR bundle.
All eight checks passed — confirming full data integrity from raw Synthea input 
through to FHIR R4 output.

```python
import pandas as pd
import json

# ── Load source files ─────────────────────────────────────────────
conditions_raw   = pd.read_csv("outputs/conditions_icd10_cleaned.csv")
encounter_merged = pd.read_csv("outputs/conditions_encounter_merged.csv")
patients         = pd.read_csv("data/patients.csv")
scored_segments  = pd.read_csv("outputs/patient_segments.csv")

with open("outputs/fhir/clinical_risk_bundle.json") as f:
    bundle = json.load(f)

# ── Extract FHIR counts ───────────────────────────────────────────
fhir_patients   = [e["resource"] for e in bundle["entry"] if e["resource"]["resourceType"] == "Patient"]
fhir_encounters = [e["resource"] for e in bundle["entry"] if e["resource"]["resourceType"] == "Encounter"]
fhir_conditions = [e["resource"] for e in bundle["entry"] if e["resource"]["resourceType"] == "Condition"]
fhir_risk       = [e["resource"] for e in bundle["entry"] if e["resource"]["resourceType"] == "RiskAssessment"]

print("══ Cross-Validation Against Source Data ══════════════")

# ── 1. Patient count ──────────────────────────────────────────────
print("\n── 1. Patients ──────────────────────────────────────")
source_patients = len(scored_segments)
fhir_pat_count  = len(fhir_patients)
print(f"  Scored patients (source)  : {source_patients}")
print(f"  Patients in FHIR bundle   : {fhir_pat_count}")
print(f"  Match                     : {'PASSED' if source_patients == fhir_pat_count else 'MISMATCH'}")

# ── 2. Encounter count ────────────────────────────────────────────
print("\n── 2. Encounters ────────────────────────────────────")
source_encounters = len(encounter_merged)
fhir_enc_count    = len(fhir_encounters)
print(f"  Encounter rows (source)   : {source_encounters}")
print(f"  Encounters in FHIR bundle : {fhir_enc_count}")
print(f"  Difference                : {source_encounters - fhir_enc_count}")
print(f"  Note: difference expected if Synthea duplicate ENCOUNTER IDs present")

# ── 3. Condition count ────────────────────────────────────────────
print("\n── 3. Conditions ────────────────────────────────────")
source_conditions = len(conditions_raw)
fhir_cond_count   = len(fhir_conditions)
print(f"  Condition rows (source)   : {source_conditions}")
print(f"  Conditions in FHIR bundle : {fhir_cond_count}")
print(f"  Match                     : {'PASSED' if source_conditions == fhir_cond_count else 'MISMATCH'}")

# ── 4. RiskAssessment count ───────────────────────────────────────
print("\n── 4. RiskAssessments ───────────────────────────────")
print(f"  Scored patients (source)       : {source_patients}")
print(f"  RiskAssessments in FHIR bundle : {len(fhir_risk)}")
print(f"  Match                          : {'PASSED' if source_patients == len(fhir_risk) else 'MISMATCH'}")

# ── 5. SNOMED code coverage ───────────────────────────────────────
print("\n── 5. SNOMED Code Coverage ──────────────────────────")
source_snomed_codes = set(conditions_raw["CODE"].astype(str).unique())
fhir_snomed_codes   = set()
for c in fhir_conditions:
    for coding in c.get("code", {}).get("coding", []):
        if coding.get("system") == "http://snomed.info/sct":
            fhir_snomed_codes.add(coding.get("code"))
print(f"  Unique SNOMED codes in source : {len(source_snomed_codes)}")
print(f"  Unique SNOMED codes in FHIR   : {len(fhir_snomed_codes)}")
missing_in_fhir = source_snomed_codes - fhir_snomed_codes
print(f"  Codes missing from FHIR       : {len(missing_in_fhir)}")
if missing_in_fhir:
    print(f"  Missing codes: {missing_in_fhir}")

# ── 6. Patient ID consistency ─────────────────────────────────────
print("\n── 6. Patient ID Consistency ────────────────────────")
source_patient_ids = set(scored_segments["PATIENT"].astype(str))
fhir_patient_ids   = set(p["id"] for p in fhir_patients)
only_in_source     = source_patient_ids - fhir_patient_ids
only_in_fhir       = fhir_patient_ids - source_patient_ids
print(f"  Patients only in source   : {len(only_in_source)}")
print(f"  Patients only in FHIR     : {len(only_in_fhir)}")
print(f"  Match                     : {'PASSED' if not only_in_source and not only_in_fhir else 'MISMATCH'}")

# ── 7. Risk score consistency ─────────────────────────────────────
print("\n── 7. Risk Score Consistency ────────────────────────")
source_scores = dict(zip(scored_segments["PATIENT"], scored_segments["risk_score"]))
fhir_scores   = {
    r["subject"]["reference"].replace("Patient/", ""): r["prediction"][0]["probabilityDecimal"]
    for r in fhir_risk
}
mismatched_scores = {
    pid: (source_scores[pid], fhir_scores[pid])
    for pid in source_scores
    if pid in fhir_scores and abs(source_scores[pid] - fhir_scores[pid]) > 0.01
}
print(f"  Mismatched risk scores    : {len(mismatched_scores)}")
print(f"  Match                     : {'PASSED' if not mismatched_scores else 'MISMATCH'}")

# ── 8. Segment consistency ────────────────────────────────────────
print("\n── 8. Segment Consistency ───────────────────────────")
source_segments = dict(zip(scored_segments["PATIENT"], scored_segments["primary_segment"]))
fhir_segments   = {}
for r in fhir_risk:
    pid  = r["subject"]["reference"].replace("Patient/", "")
    note = r.get("note", [{}])[0].get("text", "")
    for part in note.split("|"):
        if "Primary segment:" in part:
            fhir_segments[pid] = part.replace("Primary segment:", "").strip()

mismatched_segments = {
    pid: (source_segments[pid], fhir_segments.get(pid))
    for pid in source_segments
    if pid in fhir_segments and source_segments[pid] != fhir_segments.get(pid)
}
print(f"  Mismatched segments       : {len(mismatched_segments)}")
print(f"  Match                     : {'PASSED' if not mismatched_segments else 'MISMATCH'}")

print("\n══ Cross-Validation Complete ═════════════════════════")
```

## Wrting the FHIR R4 Export Module

```python
%%writefile src/build_fhir_bundle.py
import json
import os
import re
import pandas as pd
from datetime import datetime, timezone
from collections import defaultdict

DISPLAY_CORRECTIONS = {
    "N39.0":  "Urinary tract infection, site not specified",
    "E11.9":  "Type 2 diabetes mellitus : Without complications",
    "E11.2":  "Type 2 diabetes mellitus : With renal complications",
    "E11.3":  "Type 2 diabetes mellitus : With ophthalmic complications",
    "E11.4":  "Type 2 diabetes mellitus : With neurological complications",
    "O03.9":  "Spontaneous abortion : Complete or unspecified, without complication",
    "I48.9":  "Atrial fibrillation and atrial flutter, unspecified",
    "I46.9":  "Cardiac arrest, unspecified",
    "R73.0":  "Abnormal glucose tolerance test",
    "G30.9":  "Alzheimer disease, unspecified",
    "G30.0":  "Alzheimer disease with early onset",
    "C50.9":  "Breast, unspecified",
    "S62.0":  "Fracture of navicular [scaphoid] bone of hand",
    "S63.6":  "Sprain and strain of finger(s)",
    "S93.4":  "Sprain and strain of ankle",
    "S91.3":  "Open wound of other parts of foot",
    "T50.9":  "Other and unspecified drugs, medicaments and biological substances",
    "O02.0":  "Blighted ovum and nonhydatidiform mole",
    "O36.9":  "Maternal care for fetal problem, unspecified",
}

DESCRIPTION_CORRECTIONS = {
    59621000:  "Essential hypertension (disorder)",
    403190006: "Epidermal burn of skin (disorder)",
    5602001:   "Harmful pattern of use of opioid",
    201834006: "Localized, primary osteoarthritis of the hand",
}


def safe_date(dt):
    if pd.isna(dt):
        return None
    return pd.Timestamp(dt).strftime("%Y-%m-%d")


def parse_list_col(val):
    if isinstance(val, list):
        return val
    try:
        return json.loads(val.replace("'", '"'))
    except Exception:
        return []


def strip_display(desc):
    if " - " in str(desc):
        return desc.split(" - ")[0].strip()
    return desc


def get_official_display(icd10_code, current_display):
    return DISPLAY_CORRECTIONS.get(icd10_code, current_display)


def load_data(conditions_path, patients_path, scored_segments_path, map_path, universe_path):
    encounter_merged = pd.read_csv(conditions_path, parse_dates=["START", "STOP"])
    patients         = pd.read_csv(patients_path)
    scored_segments  = pd.read_csv(scored_segments_path)
    df_map           = pd.read_csv(map_path)
    universe_ref     = pd.read_csv(universe_path)

    df_map["icd10_desc"] = df_map["icd10_desc"].str.replace("—", "-", regex=False)

    for code, desc in DESCRIPTION_CORRECTIONS.items():
        universe_ref.loc[universe_ref["CODE"] == code, "DESCRIPTION"] = desc

    df_map = df_map.merge(
        universe_ref[["CODE", "DESCRIPTION"]].rename(columns={"CODE": "snomed_code"}),
        on="snomed_code",
        how="left"
    )

    print(f"[fhir_exporter] Encounter rows loaded    : {len(encounter_merged)}")
    print(f"[fhir_exporter] Patients loaded          : {len(patients)}")
    print(f"[fhir_exporter] Scored segments loaded   : {len(scored_segments)}")
    print(f"[fhir_exporter] Mapping rows loaded      : {len(df_map)}")

    return encounter_merged, patients, scored_segments, df_map


def build_patient_resource(row):
    gender_map = {"M": "male", "F": "female", "m": "male", "f": "female"}
    gender     = gender_map.get(str(row["GENDER"]).strip(), "unknown") if pd.notna(row["GENDER"]) else "unknown"
    return {
        "resourceType": "Patient",
        "id":           row["Id"],
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Patient {row['Id']} | Gender: {gender} | DOB: {safe_date(pd.to_datetime(row['BIRTHDATE'], errors='coerce'))}</div>"
        },
        "gender":    gender,
        "birthDate": safe_date(pd.to_datetime(row["BIRTHDATE"], errors="coerce")),
    }


def build_encounter_resource(enc_row):
    enc_id = f"{enc_row['ENCOUNTER']}-{pd.Timestamp(enc_row['START']).strftime('%Y%m%d')}"
    period = {"start": safe_date(enc_row["START"])}
    if pd.notna(enc_row["STOP"]):
        period["end"] = safe_date(enc_row["STOP"])
    return {
        "resourceType": "Encounter",
        "id":           enc_id,
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Encounter on {safe_date(enc_row['START'])} for patient {enc_row['PATIENT']}</div>"
        },
        "status": "finished",
        "class": {
            "system":  "http://terminology.hl7.org/CodeSystem/v3-ActCode",
            "code":    "AMB",
            "display": "ambulatory"
        },
        "subject": {"reference": f"Patient/{enc_row['PATIENT']}"},
        "period":  period
    }


def build_condition_resources(enc_row, df_map):
    snomed_codes = parse_list_col(enc_row["snomed_codes"])
    icd10_codes  = parse_list_col(enc_row["icd10_codes"])
    descriptions = parse_list_col(enc_row["descriptions"])
    conditions   = []
    enc_id       = f"{enc_row['ENCOUNTER']}-{pd.Timestamp(enc_row['START']).strftime('%Y%m%d')}"

    for i, snomed_code in enumerate(snomed_codes):
        snomed_str = str(snomed_code)
        match      = df_map[df_map["snomed_code"].astype(str) == snomed_str]

        if not match.empty:
            icd10_code  = match.iloc[0]["icd10_code"]
            icd10_desc  = get_official_display(icd10_code, strip_display(match.iloc[0]["icd10_desc"]))
            snomed_desc = match.iloc[0]["DESCRIPTION"] if "DESCRIPTION" in match.columns else (descriptions[i] if i < len(descriptions) else "Unknown")
        else:
            icd10_code  = icd10_codes[i] if i < len(icd10_codes) else "Unknown"
            icd10_desc  = get_official_display(icd10_code, "Unknown")
            snomed_desc = descriptions[i] if i < len(descriptions) else "Unknown"

        cond_id   = f"{enc_id}-{i}"
        condition = {
            "resourceType": "Condition",
            "id":           cond_id,
            "text": {
                "status": "generated",
                "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>{snomed_desc} (SNOMED: {snomed_str} | ICD-10: {icd10_code})</div>"
            },
            "subject":       {"reference": f"Patient/{enc_row['PATIENT']}"},
            "encounter":     {"reference": f"Encounter/{enc_id}"},
            "onsetDateTime": safe_date(enc_row["START"]),
            "code": {
                "coding": [
                    {
                        "system":       "http://snomed.info/sct",
                        "code":         snomed_str,
                        "display":      snomed_desc,
                        "userSelected": True
                    },
                    {
                        "system":       "http://hl7.org/fhir/sid/icd-10",
                        "code":         icd10_code,
                        "display":      icd10_desc,
                        "userSelected": False
                    }
                ],
                "text": snomed_desc
            },
            "clinicalStatus": {
                "coding": [{
                    "system":  "http://terminology.hl7.org/CodeSystem/condition-clinical",
                    "code":    "active" if pd.isna(enc_row["STOP"]) else "resolved",
                    "display": "Active" if pd.isna(enc_row["STOP"]) else "Resolved"
                }]
            }
        }
        conditions.append(condition)
    return conditions


def build_risk_assessment(patient_id, scored_row):
    return {
        "resourceType": "RiskAssessment",
        "id":           patient_id,
        "text": {
            "status": "generated",
            "div":    f"<div xmlns='http://www.w3.org/1999/xhtml'>Risk tier: {scored_row['risk_tier']} | Score: {scored_row['risk_score']} | Segment: {scored_row['primary_segment']}</div>"
        },
        "subject": {"reference": f"Patient/{patient_id}"},
        "status":  "final",
        "prediction": [
            {
                "outcome": {
                    "coding": [{
                        "system":  "http://snomed.info/sct",
                        "code":    "225338004",
                        "display": "Risk assessment"
                    }]
                },
                "probabilityDecimal": float(scored_row["risk_score"]),
            }
        ],
        "note": [
            {
                "text": (
                    f"Risk tier: {scored_row['risk_tier']} | "
                    f"Primary segment: {scored_row['primary_segment']} | "
                    f"All segments: {scored_row['all_segments']}"
                )
            }
        ]
    }


def build_bundle(encounter_merged, patients, scored_segments, df_map, output_dir="outputs"):
    os.makedirs(os.path.join(output_dir, "fhir"), exist_ok=True)

    bundle = {
        "resourceType": "Bundle",
        "id":           f"clinical-risk-bundle-{datetime.today().strftime('%Y-%m-%d')}",
        "type":         "collection",
        "timestamp":    datetime.now(timezone.utc).strftime("%Y-%m-%dT%H:%M:%S+00:00"),
        "entry":        []
    }

    scored_index   = scored_segments.set_index("PATIENT")
    enc_by_patient = encounter_merged.groupby("PATIENT")

    patient_count   = 0
    condition_count = 0
    encounter_count = 0
    risk_count      = 0

    for _, pat_row in patients.iterrows():
        patient_id = pat_row["Id"]

        if patient_id not in scored_index.index:
            continue

        scored_row = scored_index.loc[patient_id]

        bundle["entry"].append({
            "fullUrl":  f"http://clinical-risk-pipeline/Patient/{pat_row['Id']}",
            "resource": build_patient_resource(pat_row)
        })
        patient_count += 1

        if patient_id in enc_by_patient.groups:
            for _, enc_row in enc_by_patient.get_group(patient_id).iterrows():
                enc_resource = build_encounter_resource(enc_row)
                bundle["entry"].append({
                    "fullUrl":  f"http://clinical-risk-pipeline/Encounter/{enc_resource['id']}",
                    "resource": enc_resource
                })
                encounter_count += 1

                for cond in build_condition_resources(enc_row, df_map):
                    bundle["entry"].append({
                        "fullUrl":  f"http://clinical-risk-pipeline/Condition/{cond['id']}",
                        "resource": cond
                    })
                    condition_count += 1

        bundle["entry"].append({
            "fullUrl":  f"http://clinical-risk-pipeline/RiskAssessment/{patient_id}",
            "resource": build_risk_assessment(patient_id, scored_row)
        })
        risk_count += 1

    return bundle, patient_count, encounter_count, condition_count, risk_count


def run(conditions_path, patients_path, scored_segments_path, map_path, universe_path, output_dir="outputs"):
    encounter_merged, patients, scored_segments, df_map = load_data(
        conditions_path, patients_path, scored_segments_path, map_path, universe_path
    )

    bundle, patient_count, encounter_count, condition_count, risk_count = build_bundle(
        encounter_merged, patients, scored_segments, df_map, output_dir
    )

    output_path = os.path.join(output_dir, "fhir", "clinical_risk_bundle.json")
    with open(output_path, "w") as f:
        json.dump(bundle, f, indent=2, default=str, ensure_ascii=False)

    print(f"\n[fhir_exporter] ── Bundle Export ────────────────────────")
    print(f"[fhir_exporter] Output file        : {output_path}")
    print(f"[fhir_exporter] Total entries      : {len(bundle['entry'])}")
    print(f"[fhir_exporter]   Patient          : {patient_count}")
    print(f"[fhir_exporter]   Encounter        : {encounter_count}")
    print(f"[fhir_exporter]   Condition        : {condition_count}")
    print(f"[fhir_exporter]   RiskAssessment   : {risk_count}")
    print(f"[fhir_exporter] File size          : {os.path.getsize(output_path) / 1024:.1f} KB")

    return bundle
```

## Full Pipeline Control File

```python
%%writefile run_pipeline.py
import pandas as pd
import src.build_snomed_universe as step1
import src.compute_risk_scores   as step2
import src.assign_segments       as step3
import src.map_icd10             as step4
import src.build_fhir_bundle     as step5

CONDITIONS_PATH  = "data/conditions.csv"
PATIENTS_PATH    = "data/patients.csv"
MEDICATIONS_PATH = "data/medications.csv"
ENCOUNTERS_PATH  = "data/encounters.csv"
UNIVERSE_PATH    = "outputs/snomed_conditions_universe.csv"
MAP_PATH         = "data/reference/snomed_to_icd10_map.csv"
ENCOUNTER_PATH   = "outputs/intermediate/conditions_encounter_merged.csv"
SEGMENTS_PATH    = "outputs/patient_segments.csv"
OUTPUT_DIR       = "outputs"

def main():
    print("══ Clinical Risk Analysis Pipeline v2 ══════════════")

    print("\n── Sprint 1 : SNOMED Condition Mapping ───────────────")
    conditions = pd.read_csv(CONDITIONS_PATH)
    universe   = pd.read_csv(UNIVERSE_PATH)
    wide, long, merged = step1.run(
        conditions_path=CONDITIONS_PATH,
        universe_path=UNIVERSE_PATH,
        output_dir=OUTPUT_DIR
    )

    print("\n── Sprint 2 : Risk Scoring ───────────────────────────")
    patients    = pd.read_csv(PATIENTS_PATH)
    medications = pd.read_csv(MEDICATIONS_PATH)
    encounters  = pd.read_csv(ENCOUNTERS_PATH)
    scored = step2.run(
        wide=wide,
        merged=merged,
        patients=patients,
        medications=medications,
        encounters=encounters,
        output_dir=OUTPUT_DIR
    )

    print("\n── Sprint 3 : Patient Segmentation ───────────────────")
    scored, segment_summary = step3.run(
        scored=scored,
        conditions=conditions,
        universe_path=UNIVERSE_PATH,
        output_dir=OUTPUT_DIR
    )

    print("\n── Sprint 4 : ICD-10 Mapping ─────────────────────────")
    conditions_mapped, scored_with_icd10, universe = step4.run(
        conditions=conditions,
        universe=universe,
        universe_path=UNIVERSE_PATH,
        map_path=MAP_PATH,
        scored=scored,
        output_dir=OUTPUT_DIR
    )

    print("\n── Sprint 5 : FHIR R4 Bundle Export ──────────────────")
    bundle = step5.run(
        conditions_path=ENCOUNTER_PATH,
        patients_path=PATIENTS_PATH,
        scored_segments_path=SEGMENTS_PATH,
        map_path=MAP_PATH,
        universe_path=UNIVERSE_PATH,
        output_dir=OUTPUT_DIR
    )

    print("\n══ Pipeline complete ══════════════════════════════════")

if __name__ == "__main__":
    main()
```
